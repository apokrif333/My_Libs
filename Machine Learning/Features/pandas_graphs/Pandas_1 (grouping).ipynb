```python
# %matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import os
import sys

sys.path.append('../')
plt.style.use('ggplot')  # Красивые графики
plt.rcParams['figure.figsize'] = (20, 10)  # Размер картинок
```

```python
dir = os.path.abspath(os.curdir).replace('pandas_graphs', 'data/').replace("\\", '/')
bikes_df = pd.read_csv(dir + "bikes.csv",
                       sep=';', 
                       encoding='latin1', 
                       parse_dates=['Date'], 
                       dayfirst=True, 
                       index_col='Date')
bikes_df[:3]
```

```python
bikes_df.plot()
```

```python
bikes_df['Berri 1'].plot()
```

```python
berri_bikes = bikes_df[['Berri 1']].copy()
berri_bikes.loc[:, 'weekday'] = berri_bikes.index.weekday

weekday_counts = berri_bikes.groupby('weekday').mean()
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts.plot(kind='bar')
```
