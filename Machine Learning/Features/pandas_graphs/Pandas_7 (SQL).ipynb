```python
import os
import pandas as pd
import sqlite3
```

```python
# Считываение и установка индексов для SQL файлов
dir = os.path.abspath(os.curdir).replace('pandas_graphs', 'data/').replace("\\", '/')
connect = sqlite3.connect(dir + 'weather_2012.sqlite')
weather_df = pd.read_sql('SELECT * from weather_2012 LIMIT 310', connect, index_col=['id', 'date_time'])
weather_df
```

```python
# Создаём файл test_db, считываем weather_2012 и если weather_2012 существует, очищаем test_db и записываем туда weather_2012
weather_df = pd.read_csv(dir + 'weather_2012.csv')
connect = sqlite3.connect(dir + 'test_db.sqlite')
connect.execute('DROP TABLE IF EXISTS weather_2012')
weather_df.to_sql('weather_2012', connect)
```

```python
# Считываем ранее созданный файл и сортируем языком SQL
connect = sqlite3.connect('data/test_db.sqlite')
test1_df = pd.read_sql('SELECT * from weather_2012', connect)
test2_df = pd.read_sql('SELECT *from weather_2012 ORDER BY Weather', connect)
test2_df
```
