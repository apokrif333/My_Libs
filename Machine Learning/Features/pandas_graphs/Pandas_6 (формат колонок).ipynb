```python
%matplotlib inline

import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
dir = os.path.abspath(os.curdir).replace('pandas_graphs', 'data/').replace("\\", '/')
popcon_df = pd.read_csv(dir + 'popularity_contest', sep=' ')[:-1]
popcon_df.columns = ['atime', 'ctime', 'package-name', 'mru-program', 'tag']
popcon_df
```

```python
# Преобразовываем Unix-значения во время
popcon_df['atime'] = popcon_df['atime'].astype(int)
popcon_df['ctime'] = popcon_df['ctime'].astype(int)

popcon_df['atime'] = pd.to_datetime(popcon_df['atime'], unit='s')
popcon_df['ctime'] = pd.to_datetime(popcon_df['ctime'], unit='s')

popcon_df = popcon_df[popcon_df['atime'] > '1970-01-01']
```

```python
# Получаем список строк, которые не содержат lib, путём поиска lib, создания Series и инвертирования её
nonlibraries = popcon_df[~popcon_df['package-name'].str.contains('lib')]
nonlibraries.sort_values('ctime', ascending=False)
```
