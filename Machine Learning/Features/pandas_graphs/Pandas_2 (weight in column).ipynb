```python
%matplotlib inline

import os
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
# Считываем файл, не взирая на ошибки
dir = os.path.abspath(os.curdir).replace('pandas_graphs', 'data/').replace("\\", '/')
complaints_df = pd.read_csv(dir + '311_service_requests.csv')
```

```python
# Все события
all_types = complaints_df['Complaint Type'].unique()
all_types
```

```python
# Все районы
all_borough = complaints_df['Borough'].unique()
all_borough
```

```python
# Определённые события в %
all_complaints = len(complaints_df)
noise_list = ['Noise - Street/Sidewalk', 'Noise - Commercial', 'Noise - Vehicle', 
              'Noise - House of Worship', 'Noise - Park', 'Noise'
             ]

len(complaints_df.loc[complaints_df['Complaint Type'].isin(noise_list)]) / all_complaints * 100
```

```python
# Расчёт % каждого коэффициента
for c in all_types:
    print(c)
    print(len(complaints_df.loc[complaints_df['Complaint Type'] == c]) / all_complaints * 100)
```

```python
complaints_counts = complaints_df['Complaint Type'].value_counts()
complaints_counts[:10].plot(kind='bar')
```
